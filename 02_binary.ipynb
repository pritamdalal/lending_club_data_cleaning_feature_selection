# Isolate Binary Outcomes

The only purpose of this notebook is to isolate the loans that have a `loan_status` value of `Fully Paid` or `Charged Off` so as to turn this into a binary classification problem.

### Load Packages

```python
import pandas as pd
pd.options.display.max_rows = 200
```

### Read-In Data

```python
#df_default = pd.read_csv('accepted_2007_to_2018q4/accepted_2007_to_2018Q4.csv', low_memory=False)
df_default = pd.read_csv("data_reduced/lending_club_reduced.csv")
df_default
```

```python
df_default.head(1).T
```

### Explore Data

The loan status values that seem to be relevant to us are `Fully Paid` and `Charged Off`.

```python
df_default["loan_status"].value_counts()
```

### Separate Out Binary Outcomes

Let's keep only the rows with those values and discard the rest.

```python
df_binary = df_default.query('loan_status == "Fully Paid" | loan_status == "Charged Off"')
df_binary
```

### Write to CSV

```python
df_binary.to_csv('data_processed/01_binary.csv', index=False)
```
