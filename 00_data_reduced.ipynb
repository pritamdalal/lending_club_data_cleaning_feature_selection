# Data Preparation

The lending club data is pretty large, and this causes problems for a lot of the students so I am going to reduce the size of the data set to make sure this isn't a problem moving forward.

## Importing Packages

```python
import pandas as pd
```

## Reading-In Data

```python
df_lending_club_raw = pd.read_csv("accepted_2007_to_2018q4/accepted_2007_to_2018Q4.csv")
df_lending_club_raw
```

## Sampling A Million Rows of the Data

```python
df_lending_club_reduced = df_lending_club_raw.sample(1000000, random_state=0).reset_index(drop=True)
df_lending_club_reduced
```

## Exporting Reduced Data to CSV

```python
df_lending_club_reduced.to_csv("lending_club_reduced.csv", index=False)
```
